```python
# 2. find the second largest value
def second_largest(arr):
    first=second = float('-inf')
    for num in arr:
        if num > first:
            second = first
            first = num
        elif num > second and num!=first:
            second = num
    return second
arr = [10, 20, 4, -1,45, 99,]
print(second_largest(arr))
```

```python
# 3. check waether the given array is sorted or not
def check_array_is_sorted(arr):
    for i in range(1,len(arr)):
        if arr[i] < arr[i-1]:
            return False
    return True

arr = [10, 20, 4, -1,45, 99,]
# arr = [2, 2, 4, 4,5, 99,]
print(check_array_is_sorted(arr))
```

```python
# 4. Find the duplicates in given array and return the count of unique elements
# Two-Pointers apparoach
def find_duplicates_in_array(arr):
    count = 0
    for i in range(1,len(arr)):
        if arr[i] == arr[i-1]:
            count+=1
    return count

arr = [1,1,2,2,2,3,3,4,5,6,6]
print(find_duplicates_in_array(arr))
```

```python
def find_duplicates_in_array(arr):
    dup = set()
    seen = set()
    for num in arr:
        if num in seen:
            dup.add(num)
        else:
            seen.add(num)
    return (f" dup {dup} count {len(seen)}")

arr = [1,1,2,2,2,3,3,4,5,6,6]
print(find_duplicates_in_array(arr))
```

```python
# 5. Left rotate array by one place
def left_rotate_array_by_one_decimal(arr):
    temp = arr[0]
    n = len(arr)
    for i in range(1,n):
        arr[i-1] = arr[i]
    arr[n-1]=temp
    return arr
    
arr = [1,2,3,4,5]
print(left_rotate_array_by_one_decimal(arr))
```

```python
# 6. Rotate an array with d places
def rotate_array(arr,d):
    n = len(arr) # 7
    d = d%n # 3
    temp = arr[:d] # [1,2,3]
    # shifting
    # taking range from (3,7) so here 3rd index value goes in 1st index values(3,4,5,6) arr[i-d] = [3-3] = 0(here value is 4 ) and [4-3] = 1(value 5 ) and [5-3]=2(value 6) and [6-3]=3(value 7) range is completed
    for i in range(d,n):
        arr[i-d] = arr[i]
    # arr[7-3+1]=arr[4] = temp[0]  → put 1 here
    # arr[5] = temp[1]  → put 2 here
    # arr[6] = temp[2]  → put 3 here
    for i in range(len(temp)):
        arr[n-d+i]=temp[i]
    return arr

arr = [1,2,3,4,5,6,7]
d = 3
print(rotate_array(arr,d))
```

```python
# 7. Move all zeros at end
def move_zeros_end(arr):
    j = 0
    for i in range(len(arr)):
        if arr[i]!=0:
            arr[j],arr[i]=arr[i],arr[j]
            j+=1
    return arr

arr = [1,2,0,4,0,12,0,0,5,6,0,9]
print(move_zeros_end(arr))
```

```python
# 8 inding the union and intersection of two sorted arrays
def union_array(arr1,arr2):
    unique = set()
    for i in arr1:
        unique.add(i)
    for j in arr2:
        unique.add(j)
    return list(unique)

arr1 = [1,2,3,4,5]
arr2 = [3,4,5,6]
print(union_array(arr1,arr2))
```

```python
def union_array(arr1,arr2):
    res = []
    i = 0
    j = 0
    while i < len(arr1) and j < len(arr2):
        if arr1[i] < arr2[j]:
            if arr1[i] not in res:
                res.append(arr1[i])
            i+=1
        elif arr1[i] > arr2[j]:
            if arr2[j] not in res:
                res.append(arr2[j])
            j+=1
        else:
            if arr1[i] not in res:
                res.append(arr1[i])
            i+=1
            j+=1
    while i < len(arr1):
        res.append(arr1[i])
        i+=1
    while j < len(arr2):
        res.append(arr2[j])
        j+=1
                
    return res
arr1 = [1,2,3,4,5]
arr2 = [3,4,5,6]
print(union_array(arr1,arr2))
```

```python
# 10. Interection elements form two sorted arrays
def intersection(arr1, arr2):
    i=0
    j=0
    res = []
    while i < len(arr1) and j < len(arr2):
        if arr1[i]==arr2[j]:
            if arr1[i] not in res:
                res.append(arr1[i])
            i+=1
            j+=1
        elif arr1[i] <arr2[j]:
            i+=1
        else:
            j+=1
    return res
            
     
arr1 = [1, 2, 3, 4, 5]
arr2 = [3, 4, 5, 6]

print(intersection(arr1, arr2))
```

```python
def missing_number(arr):
    for i in range(1,max(arr)):
        if i not in arr:
            return i

arr = [1,2,4,5]
print(missing_number(arr))
```
